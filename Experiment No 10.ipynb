In Natural Language Processing (NLP), frequency distribution refers to the count of occurrences of each word (or token) in a given text or corpus. This helps in understanding word importance, detecting patterns, and performing tasks like language modeling, text summarization, and sentiment analysis.

##### Key Concepts:

1. **Tokenization**: The process of splitting text into individual components (words, punctuation, etc.).
2. **Stopwords**: Common words such as "the," "is," "in," which may not carry much semantic meaning but often dominate frequency distribution.
3. **Lemmatization/Stemming**: Reducing words to their base or root form, helping to group similar words (e.g., "running" â†’ "run").

##### Applications of Frequency Distribution in NLP:

- **Text Summarization**: Identify frequently occurring words that may represent the main theme of a text.
- **Sentiment Analysis**: Detect common words with positive or negative connotations.
- **Topic Modeling**: Uncover common words that may point to topics discussed in the text.


```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from collections import Counter
import matplotlib.pyplot as plt

# Ensure necessary NLTK packages are downloaded
nltk.download('punkt')
nltk.download('stopwords')

# Sample text corpus (can be replaced with any text dataset)
text = """
Natural Language Processing (NLP) is a sub-field of artificial intelligence (AI) that focuses on the interaction between computers and humans through natural language.
The ultimate objective of NLP is to enable computers to understand, interpret, and respond to human language in a way that is both valuable and meaningful.
"""

# Tokenize the text
tokens = word_tokenize(text.lower())

# Remove punctuation and stopwords
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word.isalnum() and word not in stop_words]

# Calculate the frequency distribution
def calculate_word_frequency(tokens):
    frequency = Counter(tokens)
    return frequency

# Plot the frequency distribution
def plot_frequency(frequency):
    words = list(frequency.keys())
    counts = list(frequency.values())
    
    plt.figure(figsize=(10,6))
    plt.bar(words, counts, color='green')
    plt.xlabel('Words')
    plt.ylabel('Frequency')
    plt.title('Word Frequency Distribution in Text')
    plt.xticks(rotation=45)
    plt.show()

# Get word frequency distribution
word_frequency = calculate_word_frequency(filtered_tokens)

# Display word frequency distribution
print("Word Frequency Distribution: ", word_frequency)

# Plotting the word frequency distribution
plot_frequency(word_frequency)
```

##### Explanation of Code:

**1. Importing Required Libraries**:

- `nltk` is used for tokenization, stopwords, and general text processing.
- `matplotlib.pyplot` is used for plotting the frequency distribution.
- `collections.Counter` is used to count the occurrences of each word in the corpus.

**2. Text Preprocessing**:

- Tokenization: `The word_tokenize()` function splits the raw text into individual words and punctuation.
- Lowercasing: Converts all tokens to lowercase to avoid case-sensitive discrepancies.
- Stopword Removal: Common English stopwords are filtered out using NLTKâ€™s stopword list.
- Punctuation Removal: Words that are not alphanumeric are discarded to focus on meaningful tokens.

**3. calculate_word_frequency Function**:

- This function uses `Counter` to count the occurrences of each word in the cleaned token list (filtered tokens).
- The result is a dictionary where keys are the words, and values are the corresponding frequencies.

**4. plot_frequency Function**:

- This function visualizes the word frequency distribution using a bar chart. Words are displayed on the x-axis and their frequencies on the y-axis.
- The chart is customized to make it easy to interpret the most frequent words in the corpus.

**5. Execution**:

- The calculated frequency distribution is printed, showing the most common words.
- The word distribution is then visualized to highlight the most frequently occurring words.
